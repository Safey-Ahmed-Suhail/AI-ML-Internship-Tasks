```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
print("Shape of dataset:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nFirst 5 rows:\n", df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Boxplot of Sepal Length")
plt.show()
```

**Iris Dataset Analysis:**

**Objective**
To explore and visualize the Iris dataset using pandas, seaborn, and matplotlib.

**Dataset**
The Iris dataset contains measurements of flowers like sepal length, petal length, etc.

** Observations**
- Setosa species has smaller petal length
- Virginica has larger values
- No major missing data found
