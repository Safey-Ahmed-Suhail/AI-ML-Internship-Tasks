```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
```

```python
labels = ["Billing", "Technical Issue", "Account", "General Inquiry"]
```

```python
tickets = [
    "My internet is not working properly",
    "I want a refund for my order",
    "The app keeps crashing when I open it",
    "Payment failed but money was deducted"
]
```

```python
for ticket in tickets:
    result = classifier(ticket, labels)

    print("Ticket:", ticket)
    print("Top 3 Tags:", result['labels'][:3])
    print("-"*40)
```

```python
labels = [
    "Billing Issue (refund, payment)",
    "Technical Issue (bug, crash, error)",
    "Account Issue (login, signup)",
    "General Inquiry (information)"
]
```
