```python
from transformers import pipeline

chatbot = pipeline("text-generation", model="gpt2")
```

```python
chat_history = ""

def chat(user_input):
    global chat_history

    chat_history += f"User: {user_input}\n"

    prompt = chat_history + "Bot:"

    response = chatbot(prompt, max_length=100, num_return_sequences=1)

    answer = response[0]['generated_text'].split("Bot:")[-1]

    chat_history += f"Bot: {answer}\n"

    return answer
```

```python
print(chat("Hello, who are you?"))
print(chat("What can you do?"))
print(chat("Explain AI in simple words"))
```
